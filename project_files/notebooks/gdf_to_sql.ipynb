```python
import geopandas as gpd
import pandas as pd
import sqlite3
```

```python
# importing the polygon data from the geodataframe
pasture_gdf = gpd.GeoDataFrame.from_file('../KMZs/votm/farm_simple.geojson')
pasture_gdf.head()
```

```python
#stepping through all the polygons computing the centroid and pulling the lat / lon into separate columns
pasture_gdf['centroid_lat'] = pasture_gdf['geometry'].centroid.x
pasture_gdf['centroid_lon'] = pasture_gdf['geometry'].centroid.y
```

```python
pasture_gdf.head()
```

```python
simple_df = pasture_gdf[['name', 'centroid_lat', 'centroid_lon']]
simple_df
```

```python
conn = sqlite3.connect('../db/geometry_centroids.db')
c = conn.cursor()
```

```python
c.execute('pragma encoding=UTF8')
```

```python
simple_df.to_sql('centroids', conn, index=False)
```

```python
conn.close()
```
